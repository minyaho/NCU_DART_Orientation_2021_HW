```python
# 載入生成資料
import json
with open('data.json') as jsonfile:
    data = json.load(jsonfile)
print(len(data))
```

```python
# 設定資料 mapping
mappings = {
    "settings": {
        "number_of_shards": 2,
        "number_of_replicas": 1
    },
    "mappings":{
        "properties": {
            "id": {
                "type": "integer"
            },
            "first_name": {
                "type": "keyword"
            },
            "last_name": {
                "type": "keyword"
            },
            "email": {
                "type": "keyword"
            },
            "gender": {
                "type": "keyword"
            },
            "ip_address": {
                "type": "keyword"
            },
        }
    }
        
}
```

```python
# Insert document
from elasticsearch import Elasticsearch

es = Elasticsearch(hosts='127.0.0.1', port=9200)
# 設定mapping
es.indices.create(index="hw_week10", body=mappings)
for i, doc in enumerate(data, 1):
    # index一筆document
    res = es.index(index="hw_week10", id=i, body=doc)
```

```python
# Query docs
query = {
    "query": {
        "match": {
            "gender": "Male" ,
        }
    }
}

res = es.search(index="hw_week10", body=query)
print("Got %d Hits:" % res['hits']['total']['value'])
for hit in res['hits']['hits']:
    print(hit["_source"])
```

```python
from elasticsearch.helpers import scan

a = scan(es,
    query=query,
    index="hw_week10",
)

for i,k in enumerate(a,1):
    print(i,k)
```
